### Задание 3. 
В таблицу гена, полученную по итогам практического задания первого урока, добавьте следующую информацию: количество rs для избранного гена.

```python
genes_df = pd.read_csv('genes2.csv')
genes_df.T
```

```python
#Найдем общее колиечство снипов для каждого из генов в нашей таблице. 
def find_snp(df):
    snp = {}
    for i in range(len(df)):
        handle = Entrez.esearch(db="snp", term=f"{df.loc[i]['gen']}[All Fields] AND pathogenic[Clinical_Significance] AND missense variant[Function_Class]",
                          usehistory="y")
        record = Entrez.read(handle)
        snp[df.loc[i]['gen']]=record['IdList']
    return snp

snp = find_snp(genes_df)
```

```python
#добавим снипы и их количество в датафрейм
genes_df=pd.concat([genes_df, pd.DataFrame({'snips_count' : [len(v) for v in snp.values()], 'snp' : [v for v in snp.values()]})], axis=1)
genes_df.T
```

```python
#Рассмотрим певый ген
gene_790  =Seq(genes_df.loc[0]['gene_seq'])
gene_790
```

```python
#Получим его мРНЛ
m_rna = gene_790.transcribe()
m_rna
```

```python
#Скопируем с сайта последовательность для петвого CDS (не смогла скачать) 
exons_str = '184..265,509..648,3850..3979,4826..4968,5151..5292, 5498..5669,6195..6380,6549..6661,6977..7122,7377..7508,\
7642..7875,8337..8558,8763..8951,9157..9281,9464..9594,\
14100..14212,14612..14856,15080..15326,15674..15772,\
15944..16168,16258..16440,16640..16858,17150..17317,\
17877..18059,18168..18272,18916..19155,19381..19462,\
20011..20177,20350..20514,20688..20819,21063..21264,\
21674..21864,21963..22103,22329..22430,22897..22993,\
23199..23249,23543..23587,23690..23859,23954..24028,\
24543..24755,24934..25059,25252..25407,25504..25605,\
25830..25924,26048..26150'
exons = [e.split('..') for e in exons_str.split(',')]
```

```python
#вырежем для последовательности экзоны и сошьем его в созревшую мРНК
processed_m_rna = ''
for exon in exons:
    processed_m_rna+=(str(m_rna[int(exon[0])-1:int(exon[1])]))
processed_m_rna=Seq(processed_m_rna)
processed_m_rna
```

```python
protein = processed_m_rna.translate()
len(protein), protein
```

1. Синонимичная замена (silent)

```python
#Сделаем единичную замену третьего основания третьего кодона processed_m_rna - Синонимичная замена (silent)
mutated_m_rna1 = processed_m_rna[0:8]+'A'+processed_m_rna[9:]
mutated_m_rna1
```

```python
#С мутировавшей мРНК транслируем белок 
mutated_protein1 = mutated_m_rna1.translate()
mutated_protein1
```

2. Несинонимичная замена (nonsense)

```python
#Сделаем единичную замену второго основания 10 кодона processed_m_rna, вместо UCG мы получим стоп кодон UAG
mutated_m_rna2 = processed_m_rna[0:28]+'A'+processed_m_rna[29:]
mutated_m_rna2
```

```python
#С мутировавшей мРНК транслируем белок 
mutated_protein2 = mutated_m_rna2.translate(to_stop=True)
mutated_protein2
```
