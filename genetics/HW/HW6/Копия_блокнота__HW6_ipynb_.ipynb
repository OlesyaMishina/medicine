```python
#скачаем риды
!curl -L ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR129/011/SRR12905311/SRR12905311_1.fastq.gz -o ./sample_data/SRR12905311_1.fastq.gz
!curl -L ftp://ftp.sra.ebi.ac.uk/vol1/fastq/SRR129/011/SRR12905311/SRR12905311_2.fastq.gz -o ./sample_data/SRR12905311_2.fastq.gz
```

```python
#оценим качество прочтений
!fastqc ./sample_data/SRR12905311_1.fastq.gz
```

```python
!fastqc ./sample_data/SRR12905311_2.fastq.gz
```

```python
#Попробуем триммировать риды
!fastp --detect_adapter_for_pe \
      --overrepresentation_analysis \
      --correction --cut_right --thread 2 \
      --html sample_data/anc.fastp.html \
      --json sample_data/anc.fastp.json \
      -i sample_data/SRR12905311_1.fastq.gz -I sample_data/SRR12905311_2.fastq.gz \
      -o sample_data/trimmed_SRR1.fastq.gz -O sample_data/trimmed_SRR2.fastq.gz
```

```python
!fastqc ./sample_data/trimmed_SRR1.fastq.gz
!fastqc ./sample_data/trimmed_SRR2.fastq.gz
```

Триммирование не дало особенных результатов, чего и словедовало ожидать, так как изначальные риды уже триммированы. Скачаем геном и сделаем выравнивание.

```python
!efetch -db nuccore -id 	NC_002516.2 -format FASTA > sample_data/referens.fasta
```

```python
#проиндексируем геном
!bwa index sample_data/referens.fasta
```

```python
#и выравнивание их на геном, сразу переведя файл в формат bam и отсортировав геном по расположению
!bwa mem sample_data/referens.fasta sample_data/SRR12905311_1.fastq.gz sample_data/SRR12905311_2.fastq.gz | samtools view -bS | samtools sort -o aln.bam
```

```python
#можно посмотреть заголовок файла
!samtools view -h sample_data/aln.bam | head
```

```python
#проидексируем файл bam для более быстрого доступа
!samtools index sample_data/aln.bam
```

```python
# и посмотрим статистику
!samtools stats sample_data/aln.bam | head -n30
```

Mожно посчитать качество выравнивания как = (mapped and paired - reads MQ0)/raw total sequences=(11166604-107553)/12742882=0,87. Согласно статистике качество выравнивания достаточно хорошее.

```python
#проидексируем файл референсного генома
!samtools faidx sample_data/referens.fasta
```

```python
#Для выявления мутаций сначала получим pileup-файл, который суммирует все найденные отличия от референсного геноиа
!samtools mpileup -B -f sample_data/referens.fasta sample_data/aln.bam > sample_data/snp.mpileup
```

```python
#запустим varscan для поиска snp
!varscan mpileup2snp sample_data/snp.mpileup > sample_data/snp.txt
```

```python
!cat sample_data/snp.txt | head -n30
```

Первые 12 snp относятся к одному гену, dnaA chromosome replication initiator DnaA - инициатор репликации хромосомы ДНКА. Это белковый комплекс, названный реплисомой. Он узнаёт точку инициации и запускает механизм репликации. Соответвенно нарушения в этом гене могут привести к нарушению синтеза реплисомы и нарушению репликации ДНК.

Следующие 9 snp лежат в гене dnaN DNA polymerase III subunit beta, который учавствует трансляции бета-субъединицы ДНК-полимеразы III ( это компонент скользящего зажима, который участвует в репликации прокариотической ДНК ), то есть нарушения в этом гене так же  могут привести к нарушению репликации ДНК.

Следующим идет ген rec FDNA replication and repair protein RecF он отвечает за синтез белка репликации и репарации ДНК, мутации я нет могут приводить еще и к нарушению процесса репарации ДНК.
