–¢—Ä–∏–º–º–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–µ –¥–∞–ª–æ –æ—Å–æ–±–µ–Ω–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, —á–µ–≥–æ –∏ —Å–ª–æ–≤–µ–¥–æ–≤–∞–ª–æ –æ–∂–∏–¥–∞—Ç—å, —Ç–∞–∫ –∫–∞–∫ –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–µ —Ä–∏–¥—ã —É–∂–µ —Ç—Ä–∏–º–º–∏—Ä–æ–≤–∞–Ω—ã. –°–∫–∞—á–∞–µ–º –≥–µ–Ω–æ–º –∏ —Å–¥–µ–ª–∞–µ–º –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ.

```python
!efetch -db nuccore -id 	NC_002516.2 -format FASTA > sample_data/referens.fasta
```

```python
#–ø—Ä–æ–∏–Ω–¥–µ–∫—Å–∏—Ä—É–µ–º –≥–µ–Ω–æ–º
!bwa index sample_data/referens.fasta
```

```python
#–∏ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏–µ –∏—Ö –Ω–∞ –≥–µ–Ω–æ–º, —Å—Ä–∞–∑—É –ø–µ—Ä–µ–≤–µ–¥—è —Ñ–∞–π–ª –≤ —Ñ–æ—Ä–º–∞—Ç bam –∏ –æ—Ç—Å–æ—Ä—Ç–∏—Ä–æ–≤–∞–≤ –≥–µ–Ω–æ–º –ø–æ —Ä–∞—Å–ø–æ–ª–æ–∂–µ–Ω–∏—é
!bwa mem sample_data/referens.fasta sample_data/SRR12905311_1.fastq.gz sample_data/SRR12905311_2.fastq.gz | samtools view -bS | samtools sort -o aln.bam
```

```python
#–º–æ–∂–Ω–æ –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ñ–∞–π–ª–∞
!samtools view -h sample_data/aln.bam | head
```

```python
#–ø—Ä–æ–∏–¥–µ–∫—Å–∏—Ä—É–µ–º —Ñ–∞–π–ª bam –¥–ª—è –±–æ–ª–µ–µ –±—ã—Å—Ç—Ä–æ–≥–æ –¥–æ—Å—Ç—É–ø–∞
!samtools index sample_data/aln.bam
```

```python
# –∏ –ø–æ—Å–º–æ—Ç—Ä–∏–º —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫—É
!samtools stats sample_data/aln.bam | head -n30
```

M–æ–∂–Ω–æ –ø–æ—Å—á–∏—Ç–∞—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏—è –∫–∞–∫ = (mapped and paired - reads MQ0)/raw total sequences=(11166604-107553)/12742882=0,87. –°–æ–≥–ª–∞—Å–Ω–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–µ –∫–∞—á–µ—Å—Ç–≤–æ –≤—ã—Ä–∞–≤–Ω–∏–≤–∞–Ω–∏—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ö–æ—Ä–æ—à–µ–µ.

```python
#–ø—Ä–æ–∏–¥–µ–∫—Å–∏—Ä—É–µ–º —Ñ–∞–π–ª —Ä–µ—Ñ–µ—Ä–µ–Ω—Å–Ω–æ–≥–æ –≥–µ–Ω–æ–º–∞
!samtools faidx sample_data/referens.fasta
```

```python
#–î–ª—è –≤—ã—è–≤–ª–µ–Ω–∏—è –º—É—Ç–∞—Ü–∏–π —Å–Ω–∞—á–∞–ª–∞ –ø–æ–ª—É—á–∏–º pileup-—Ñ–∞–π–ª, –∫–æ—Ç–æ—Ä—ã–π —Å—É–º–º–∏—Ä—É–µ—Ç –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –æ—Ç–ª–∏—á–∏—è –æ—Ç —Ä–µ—Ñ–µ—Ä–µ–Ω—Å–Ω–æ–≥–æ –≥–µ–Ω–æ–∏–∞
!samtools mpileup -B -f sample_data/referens.fasta sample_data/aln.bam > sample_data/snp.mpileup
```

```python
#–∑–∞–ø—É—Å—Ç–∏–º varscan –¥–ª—è –ø–æ–∏—Å–∫–∞ snp
!varscan mpileup2snp sample_data/snp.mpileup > sample_data/snp.txt
```
