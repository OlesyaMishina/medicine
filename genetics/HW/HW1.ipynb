```python
from Bio import Entrez
from Bio import SeqIO
Entrez.email = "A.N.Other@example.com"
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
dis = pd.read_table('gene_disease_associations.tsv')
dis
```

```python
thalassemia = dis[dis['diseaseName'] == 'beta Thalassemia']
thalassemia
```

```python
def search_chromosome(geneId):
    url = 'https://www.ncbi.nlm.nih.gov/gene/?term='+str(geneId)
    return pd.read_html(url)[0]['Location'][0].split()[0]
```

```python
def get_info(list_of_genes):
    genes_info=[]
    for gene in list_of_genes:
        handle = Entrez.efetch(db="gene", id=gene, rettype = 'xml')
        gene_locus = ET.parse(handle).getroot().find("Entrezgene/Entrezgene_locus")
        region = gene_locus.find("Gene-commentary/Gene-commentary_seqs/Seq-loc/Seq-loc_int/Seq-interval")
        start = int(region.find("Seq-interval_from").text) + 1
        end = int(region.find("Seq-interval_to").text) + 1
        gi_id = region.find("Seq-interval_id/Seq-id/Seq-id_gi").text
        strand = region.find("Seq-interval_strand/Na-strand").get("value")
        nucleotide = search_chromosome(str(gene))
        genes_info.append({'gen' : gene, 'gene_id' :  gi_id, 'nucleotide' : nucleotide, 'start' : start, 'end' : end, 'strand' : strand,})
    return genes_info
```

```python
genes_df = pd.DataFrame(get_info(thalassemia.geneId))
```

```python
genes_df
```

```python
gene_lenght=[]
for i in range(len(genes_df)):
    nuc_id = genes_df.loc[i]['nucleotide']
    handle = Entrez.efetch(db="nucleotide", id=nuc_id, rettype = 'fasta', retmode = 'text')
    chromosome = SeqIO.read(handle, 'fasta')
    print(genes_df.loc[i]['gen'])
    print(chromosome.description)
    seq_gene = chromosome.seq[int(genes_df.loc[i]['start']):int(genes_df.loc[i]['end'])]
    gene_lenght.append(len(seq_gene))
    if genes_df.loc[i]['strand']=='plus':
        seq_gene = chromosome.seq[int(genes_df.loc[i]['start']):int(genes_df.loc[i]['end'])]
        print(seq_gene)
    else:
        print(seq_gene[::-1])
    
```
