```python
#Найдем общее колиечство транскриптов и белков для каждого из генов в нашей таблице. 
#Скачаем все транскрипты генов базы, возьмем только мРНК
def find_trancripts(df):
    trancripts = {}
    for i in range(len(df)):
        handle = Entrez.esearch(db="nucleotide", term=f"{df.loc[i]['gen']}[Gene Name] AND {df.loc[i]['geneSymbol']}[entrezID] AND human[Organism] AND RefSEq[KeyWord]", idtype='acc')
        record = Entrez.read(handle)
        answer = [s for s in record['IdList'] if s[1]=='M']
        trancripts[df.loc[i]['gen']]=answer
    return trancripts

trancripts = find_trancripts(gene_df)
```

```python
#добавим транскрипты и их количество в датафрейм
genes_df=pd.concat([genes_df, pd.DataFrame({'trancripts_count' : [len(v) for v in trancripts.values()], 'trancriptr' : [v for v in trancripts.values()]})], axis=1)
genes_df
```

```python
#Скачаем все белки, транслируемые с генов из базы
def find_proteins(df):
    proteins = {}
    for i in range(len(df)):
        handle = Entrez.esearch(db="protein", term=f"{df.loc[i]['geneSymbol']}[Gene Name] AND {df.loc[i]['gen']}[entrezID] AND human[Organism] AND RefSEq[KeyWord]", idtype='acc')
        answer = Entrez.read(handle)
        # answer = [s for s in record['IdList'] if s[1]=='M']
        proteins[df.loc[i]['gen']]=answer['IdList']
    return proteins

proteins = find_proteins(gene_df)
```
